```python
import pandas as pd
from datetime import datetime

what     = str(input('what food? :'))
how_many = float(input('how_many?:'))
li       = []
if what not in foods.keys():
    print('sorry we dont have that food yet, maybe you should add it')

else:
    di = {}
    di['nutrient'] = 'food'
    di['amount']   = what
    li.append(di)
    for key in foods[what].keys():
        di             = {}
        # you can throw an if keys == 0: if you dont want to imprint everytime
        #di['food']     = what
        di['nutrient'] = key
        di['amount'  ] = foods[what][key] * how_many
        
        li.append(di)
        
df = pd.DataFrame(li).set_index('nutrient')

print(list(li))
df
```
